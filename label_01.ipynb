```python
import pandas as pd
```

```python
df = pd.read_csv('/content/labeledlyricsallfixed.csv')
#df = df0['explicit_score']
df
```

```python
label = []

for ind in df.index:
    score = df['explicit_score'][ind]
    if score <= 0.55:
        lab = 0
    else:
        lab = 1
    label.append(lab)
```

```python
# Create the 'label' column based on the condition
df['label'] = df['explicit_score'].apply(lambda score: 0 if score <= 0.55 else 1)

# Create 'explicit' and 'non-explicit' columns based on the 'label' column
df['explicit'] = df['label'].apply(lambda label: 1 if label == 1 else 0)
df['non-explicit'] = df['label'].apply(lambda label: 1 if label == 0 else 0)

# Drop the 'label' column if you don't need it
df.drop(['label'], axis=1, inplace=True)
```

```python
df['explicit_count'] = pd.Series(label)
```
