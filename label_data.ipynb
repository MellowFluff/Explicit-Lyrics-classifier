```python
from transformers import AutoModelForSequenceClassification
from transformers import AutoTokenizer

import numpy as np
import pandas as pd
import torch
from tqdm.auto import tqdm
```

```python
#download tokenizer and model
MODEL = "cardiffnlp/twitter-roberta-base-offensive"

tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
#infer a few sentences
text = ["Hands up!"]

def label_with_big_model(text):
    encoded_input = tokenizer(text, return_tensors='pt', padding=True)
    output = model(**encoded_input)
    scores = torch.softmax(output[0],1).detach().numpy()
    return scores #return as array([non-explicit score, explicit score]); non-explicit score + explicit score = 1.

label_with_big_model(text)
```

```python
task='offensive'
MODEL = f"cardiffnlp/twitter-roberta-base-{task}"

tokenizer = AutoTokenizer.from_pretrained(MODEL)
```

```python
df2 = pd.read_csv('/content/lyriconlytest.csv')
#df2 = df.head(100)
df2
```

```python
score_list = []
total_rows = df2.shape[0]
batch_size = 100

for i in tqdm(range(total_rows//batch_size)):
    scores = label_with_big_model(df2.iloc[i*batch_size:(i+1)*batch_size,0].tolist())[:,1]
    score_list+=list(scores)
```

```python
score_list
```

```python
df2['explicit_score'] = pd.Series(score_list)
```
