```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/trainingdatacomplete.csv')
df
```

```python
#label x and y

x = df['lyrics']
y = df['explicit_count']
```

```python
#split data 20%

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
from sklearn import utils

label_transformed = LabelEncoder()
y_transformed = label_transformed.fit_transform(y)

x_train, x_test, y_train_transformed, y_test_transformed = train_test_split(x, y_transformed, test_size=0.2)
```

```python
x_test
```

```python
#ใช้ classifier LinerSVC

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.svm import LinearSVC

Model = Pipeline([
    ('vect', CountVectorizer()),
    ('tfidf', TfidfTransformer()),
    ('clf', LinearSVC()),
    ])

#train model
Model.fit(x_train, y_train_transformed)
```

```python
#test set

label_mapping = {0: 'non-explicit', 1: 'explicit'}
pred_labels_transformed = Model.predict(x_test)

print(x_test)
print(pred_labels_transformed)

pred_labels = label_transformed.inverse_transform(pred_labels_transformed)

testset_df = pd.DataFrame({'lyrics': x_test, 'predicted_label': pred_labels})
testset_df['explicit_count'] = testset_df['predicted_label'].map(label_mapping)
testset_df.to_csv('predictiontestset.csv', index=False)

pred_df = pd.read_csv('predictiontestset.csv')
pred_df
```

```python
#check report metrics and accuracy

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
y_true = y_test_transformed
y_pred = pred_labels

target_names = ['explicit', 'non-explicit']
print(classification_report(y_true, y_pred, target_names=target_names))

confusionmatrix = confusion_matrix(y_true, y_pred)
print(confusionmatrix)

accuracy = accuracy_score(y_true, y_pred)
print(accuracy)

# precision    recall  f1-score   support

    #explicit       0.95      0.98      0.96     20291
#non-explicit       0.91      0.81      0.86      5642

    #accuracy                           0.94     25933
   #macro avg       0.93      0.89      0.91     25933
#weighted avg       0.94      0.94      0.94     25933

#[[19852   439]
 #[ 1098  4544]]
#0.9407318860139591
```

```python
#ดึงมาใช้

input = ['bitch']

pred = Model.predict(input)
pred_labels = label_transformed.inverse_transform(pred)

predict_df = pd.DataFrame({'lyrics': input, 'predicted_label': pred_labels})
predict_df['explicit_count'] = predict_df['predicted_label'].map(label_mapping)

predict_df
```

```python
def pred_lyrics(input):
    input = [input]
    pred = Model.predict(input)
    label_ex = [label_mapping[label] for label in pred]
    return label_ex
```
